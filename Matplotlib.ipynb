# Matplotlib

```python
# coding:utf-8
import numpy as np
from matplotlib import pylab as plt

x = np.linspace(0,10,20)
y = x*x + 2

# figure 相当于绘画用的画板，而 axes 则相当于铺在画板上的画布。我们将图像绘制在画布上，于是就有了 plot，set_xlabel 等操作。
fig, axes = plt.subplots()  
axes.plot(x,y,'r')
plt.show()
```
